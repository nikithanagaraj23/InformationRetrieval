```python
import os
from collections import Counter
import nltk
from nltk.tokenize import word_tokenize
import glob
import numpy as np
import pickle
```

```python
def positionIndex(word_list, docid):
        for word in word_list:
            indices = [i for i, x in enumerate(word_list) if x == word]
            gaps = [j - i for i, j in zip(indices[:-1], indices[1:])] 
            gaps = [indices[0]]+ gaps
            if (position_index_dict.get(word)) != None and (docid not in completedfiles) :
                newdata = (docid,len(gaps),gaps)
                if not newdata in position_index_dict.get(word):
                    position_index_dict.get(word).append((docid,len(gaps),gaps))

            else:
                if not position_index_dict.get(word):
                    position_index_dict[word] = [(docid,len(gaps),gaps)]
                    if docid not in visitedfiles:
                        visitedfiles.append(docid)
```

```python
position_index_dict = {}
visitedfiles = []
completedfiles = []

def createPositionIndex():
    for filename in glob.glob("corpus/*.txt"):
        fo = open(filename, "r")
        data = fo.read()
        tokens = nltk.word_tokenize(data)
        unigramlist = nltk.word_tokenize(data)
        positionIndex(unigramlist,(os.path.basename(filename).split(".")[0]))
        completedfiles.append((os.path.basename(filename).split(".")[0]))
        print(filename," done")
```

```python
def getCommonDocs(term1,term2,k):
    documentlist = []
#     print(position_index_dict[term1])
    term1_docs = []
    term2_docs = []
    for doc in position_index_dict[term1]:
        term1_docs.append(doc[0])
    for doc in position_index_dict[term2]:
        term2_docs.append(doc[0])
    commondocs = list(set(term1_docs) & set(term2_docs))   
    return proximityDecode(term1,term2,k,commondocs)
```

```python
def getDecodedPos(encoded_pos):
    return np.cumsum(encoded_pos)
```

```python
def proximityDecode(term1,term2,k,commondocs):
    term1_docs = {}
    term2_docs = {}
    for doc in position_index_dict[term1]:
        if (doc[0] in commondocs):
            term1_docs[doc[0]] = [doc[1],getDecodedPos(doc[2])]
#             term1_docs.append([doc[0],doc[1],getDecodedPos(doc[2])])
    for doc in position_index_dict[term2]:
        if (doc[0] in commondocs):
            term2_docs[doc[0]] = [doc[1],getDecodedPos(doc[2])]
#             term2_docs.append([doc[0],doc[1],getDecodedPos(doc[2])])
#     print(term1_docs["Carbon_dioxide"])
#     print(term2_docs["Carbon_dioxide"])
    return proximityQueries(term1_docs,term2_docs,k,commondocs)
```

```python
def proximityQueries(term1_docs,term2_docs,k,commondocs):
    doclist = []
    for doc in commondocs:
#         print(doc)
        term1 = term1_docs[doc][1]
        term2 = term2_docs[doc][1]
        
        if len(term1)<=len(term2):
            document_done = False
            for pos in term1:
                for val in term2:
                    if not document_done:
                        if val in range(pos-k,pos+k+1):
                            doclist.append(doc)
                            document_done = True  
        else:
            document_done = False
            for pos in term2:
                for val in term1:
                    if not document_done:
                        if val in range(pos-k,pos+k+1):
                            doclist.append(doc)
                            document_done = True  
    return doclist
```
