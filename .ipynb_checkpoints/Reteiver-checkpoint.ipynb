```python
def inverted_index_for_unigram():
    # term -> (docid,tf)
    inverted_index = {}
    # docid -> tf
    doc_id_tf_dict = {}
    #number_of_Token = {}
    total_num_of_docs = 0
    try:
        doc_count=1
        #path = os.path.join(CURRENT_DIRECTORY,'corpus')
        for filename in listdir(INPUT_DIR):
            if isfile(join(INPUT_DIR, filename)) and filename[0] != '.':
                with open(join(INPUT_DIR, filename), 'r') as f:
                    
                    print (filename)
                    document = f.read()
                    file_name = str(filename).split('html')[0]
                    print ("Generating index for " + file_name)
                    number_of_Token[file_name] = len(document.split())
                    for word in document.split():
                        if not word in inverted_index:
                            doc_id_tf_dict = {file_name : 1}
                            inverted_index[word] =  doc_id_tf_dict
                        elif file_name in inverted_index[word]:
                            doc_id_tf_dict = inverted_index[word]
                            value = doc_id_tf_dict[file_name]
                            value = value + 1
                            doc_id_tf_dict[file_name] = value
                        else:
                            doc_id_tf_dict = {file_name : 1}
                            inverted_index[word].update(doc_id_tf_dict)
                    doc_count = doc_count + 1
            f.close()
        print ("token count : " )
        print (number_of_Token)
        total_num_of_docs = doc_count-1
    except:
        print ("Error in try block of inverted_index_for_unigram!")
        print (traceback.format_exc())
    return inverted_index,total_num_of_docs
```

```python
def get_name_len(unigram_inverted_index):

    #find number of terms in entire corpus (including duplicates)
    total_corpus_freq = 0
    doc_names = set()
    doc_length =defaultdict(lambda: 0)



    for word in list(unigram_inverted_index):
        doc_count_list = unigram_inverted_index[word]

        for docName, tf_doc in doc_count_list.items():

            doc_names.add(docName)        
            doc_length[docName]+=tf_doc
            total_corpus_freq+=tf_doc

    
    return doc_names, doc_length, total_corpus_freq
```

```python
doc_names, doc_length, total_corpus_freq = get_name_len(unigram_inverted_index)
```

```python
def get_term_corpus(query):
    
    term_freq_doc = {}
    corpus_freq = {}
    
    for word in query.split():
        
        tf_counts = {}
        doc_count_list = unigram_inverted_index[word]
        word_corpus_freq = 0
        
        for docName, tf_doc in doc_count_list.items():

            tf_counts[docName] = tf_doc
            word_corpus_freq   +=tf_doc
            #i+=1
        
        term_freq_doc[word] = tf_counts
        corpus_freq[word] = word_corpus_freq
    
    return term_freq_doc, corpus_freq
```

```python
def calc_lds(query_words, doc_names, term_freq_doc, mu = 1500):

    prob_word_doc = {}
    
    for word in query_words:
        prob_doc = {}
        
        for doc in doc_names:
            
            if doc in term_freq_doc[word]:

                prob_doc[doc] = (term_freq_doc[word][doc] + (mu*corpus_freq[word]/total_corpus_freq))/(doc_length[doc] + mu)
            
            else:
                prob_doc[doc] = (mu*corpus_freq[word]/total_corpus_freq)/(doc_length[doc] + mu)
        
        prob_word_doc[word]=prob_doc

    
    
    return prob_word_doc 
```

```python
def calc_lds(query_words, doc_names, term_freq_doc, corpus_freq, total_corpus_freq, mu = 1500):

    #print(term_freq_doc.keys())
    #print(corpus_freq)
    
    prob_word_doc = pd.DataFrame()
    
    for word in query_words:
        prob_doc = {}
        
        for doc in doc_names:
            
            if doc in term_freq_doc[word]:

                prob_doc[doc] = (term_freq_doc[word][doc] + (mu*corpus_freq[word]/total_corpus_freq))/(doc_length[doc] + mu)
            
            else:
                prob_doc[doc] = (mu*corpus_freq[word]/total_corpus_freq)/(doc_length[doc] + mu)
       
            prob_word_doc.loc[doc,word]=prob_doc[doc]
        

        
    
    return prob_word_doc 
```

```python
def get_rankings(query_list, doc_names, term_freq_doc):
    
    final_result = pd.DataFrame()
    
    for i, query in enumerate(query_list):
        
        print(query)
      
        doc_names, doc_length, total_corpus_freq = get_name_len(unigram_inverted_index)
        term_freq_doc, corpus_freq               = get_term_corpus(query)
        
        query_words = query.split()
        
        prob_word_doc = calc_lds(query_words, doc_names, term_freq_doc, corpus_freq, total_corpus_freq, 1500)
        #prob_word_doc.loc[:,'LMDirichlet_score'] = np.log(prob_word_doc[prob_word_doc.columns[0]]) +
        #                                            np.log(prob_word_doc[prob_word_doc.columns[1]])
        
        if 'LMDirichlet_score' not in prob_word_doc.columns: 
            prob_word_doc.loc[:,'LMDirichlet_score'] = np.log(prob_word_doc[prob_word_doc.columns[0]]) + np.log(prob_word_doc[prob_word_doc.columns[1]])
        
        else:
            for col in prob_word_doc.columns[2:]:
                print(np.log(prob_word_doc[col]))
                prob_word_doc.loc[:,'LMDirichlet_score'] += np.log(prob_word_doc[col]) 
        
        temp = prob_word_doc[['LMDirichlet_score']].sort_values('LMDirichlet_score', ascending = False).iloc[:100]
        temp.loc[:, 'QO']          = 'Q0'
        temp.loc[:, 'query_id']    = i+1
        temp.loc[:, 'rank']        = [i for i in range(1, len(temp) + 1)]
        temp.loc[:, 'system_name'] = 'LM Dirichlet'
        
        
        final_result = final_result.append(temp)
        
        #break

    return final_result
```

```python
query_list = ['carbon emission','cutting carbon emission','flexible dieting','flexible dieting and vegetarian',
              'information on pest control in greenhouse','pest control greenhouse','greenhouse apples apple',
              'green house apples apple']
```

```python
final_result = get_rankings(query_list, doc_names, term_freq_doc)
```

```python
for qid in final_result['query_id'].unique():
    
    final_result[final_result['query_id'] == qid].to_csv('./output/QID_' + str(qid) + '_LDS.csv')
```
